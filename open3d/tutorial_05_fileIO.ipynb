### File IO
This tutorial shows how basic data structures are read and written by Open3D.
#### Point cloud
The code below reads and writes a point cloud.

```python
import open3d as o3d 
import numpy as np 
import matplotlib.pyplot as plt 
import copy
import Open3D.examples.python.open3d_tutorial as o3dtut
```

```python
print("Testing IO for point cloud ...")
pcd = o3d.io.read_point_cloud("Open3D/examples/test_data/fragment.pcd")
print(pcd)
o3d.io.write_point_cloud("copy_fragment.pcd", pcd)
o3d.visualization.draw_geometries([pcd])
```

By default, Open3D tries to infer the file type by the filename extension. The following point cloud file types are supported:

![open3d_fileformat.jpg](open3d_fileformat.jpg)

It's also possible to specify the file type explicitly. In this case, the file extension will be ignored.

```python
pcd = o3d.io.read_point_cloud("Open3D/examples/test_data/my_points.txt", format='xyz')
print("pcd : ", pcd)
print(np.asarray(pcd.points))
o3d.visualization.draw_geometries([pcd])
```

#### Mesh
The code below reads and writes a mesh.

```python
print("Testing IO for meshes...")
mesh = o3d.io.read_triangle_mesh("Open3D/examples/test_data/knot.ply")
print(mesh)
mesh.compute_vertex_normals()
o3d.io.write_triangle_mesh("copy_knot.ply", mesh)
o3d.visualization.draw_geometries([mesh])
```

```python
mesh1 = o3d.io.read_triangle_mesh("copy_knot.ply")
o3d.visualization.draw_geometries([mesh1])
```

```python
mesh1.compute_vertex_normals()
o3d.visualization.draw_geometries([mesh1])
```

```python
mesh.paint_uniform_color([1, 0.7, 0])
o3d.visualization.draw_geometries([mesh])
```

Compared to the point cloud data structure, a mesh has triangles that define the 3D surface.

By default, Open3D tries to infer the file type by the filename extension. The following mesh file types are supported:

![mesh_fileformat](mesh_fileformat.jpg)

#### Image
The code below reads and writes an image.

```python
print("Testing IO for images ...")
img = o3d.io.read_image("Open3D/examples/test_data/lena_color.jpg")
print(img)
o3d.io.write_image("copy_lena_color.jpg", img)
```

```python
o3d.visualization.draw_geometries([img])
```

The size of the image is readily displayed using **print(img)**.

Both **jpg** and **png** image files are supported.
